```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<b>1. Data Preparation:</b>

```python
array =  np.array(np.random.uniform(-5, 35, (10, 12)))
array
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', "July", 'August', 'September', 'October', 'November', 'December']
citys = ['Rosario', 'Buenos Aires', 'Cordoba', 'Santa Fe', 'Mendoza', 'Salta', 'Bariloche', 'Mar del Plata', 'Misiones', 'Nelson']
```

```python
df = pd.DataFrame(array, citys, months)
```

```python
df
```

<b>2. Data Analysis:</b>

```python
median_temperatures = df.mean(axis=1)
median_temperatures
```

```python
median_temperatures.idxmax()
```

```python
median_temperatures.idxmin()
```

<b>3. Data visualization</b>

```python
plt.figure(figsize=(20, 12))

for city, temperature in df.iterrows():
    plt.plot(temperature.index, temperature.values, label=city)

# Personalizar el gráfico
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
