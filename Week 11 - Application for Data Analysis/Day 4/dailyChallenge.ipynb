<b>Exploratory Data Analysis:</b>

Utilize Pandas to summarize key statistics (mean, median, standard deviation) for numerical columns.<br>
Explore the distribution of power plants by country and fuel type.<br>

```python
for col in df2.columns:
    if pd.api.types.is_numeric_dtype(df2[col]):
        print(f"Column: {col}")
        print(f"Mean: {df2[col].mean()}")
        print(f"Median: {df2[col].median()}")
        print(f"Standard deviation: {df2[col].std()}")
        print("\n")
```

```python
df2.groupby(['country', 'primary_fuel']).size().reset_index(name='Power plants')
```
