```python
import pandas as pd
```

```python
data = pd.read_csv('datascience_salaries.csv')
df = pd.DataFrame(data)
```

```python
#df.head()
#df.dtypes
df.info()
```

All the data is structured
Structured data allows for easy analysis with Pandas functionalities. You can perform calculations, create visualizations, and potentially build models on this data. Unstructured data requires different approaches depending on the data type (e.g., NLP for text).

```python
missing_counts = df.isnull().sum()
print(missing_counts)
```

```python
print(df.groupby('experience_level')['salary'].agg(['mean', 'median']))
```

```python
import matplotlib.pyplot as plt

aux = df.groupby('experience_level')['salary'].agg(['mean', 'median'])    
    

plt.figure(figsize=(10, 6))
aux['mean'].plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary')
plt.title('Average Salary by Experience Level')
plt.xticks(rotation=0) 
plt.tight_layout() 

# Display the chart
plt.show()
```

```python
plt.figure(figsize=(10, 6))
aux['median'].plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary')
plt.title('Average Salary by Experience Level')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust spacing between elements

# Display the chart
plt.show()
```
