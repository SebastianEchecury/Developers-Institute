```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

<b>Exercise 1

```python
data = pd.read_csv('Student Mental health.csv')
df = pd.DataFrame(data)
df.head()
```

```python
sns.histplot(df['What is your CGPA?'], bins=10, color='skyblue', kde=True)
plt.title('Histogram of CGPA')
plt.show()
```

<b>Exercise 2

```python
#sns.barplot(x='Choose your gender', y='Do you have Anxiety?', data=df)
sns.countplot(x='Do you have Anxiety?', hue='Choose your gender', data=df)
plt.title('Anxiety Rate by Gender')
plt.show()
```

<b>Exercise 3

```python

sns.lmplot(x='Age', y='Panic Attack (Yes=1, No=0)', data=df)

# Adding plot title and labels
plt.title('Scatter Plot with Regression Line')
plt.xlabel('Age')
plt.ylabel('Panic Attack (Yes=1, No=0)')

# Displaying the plot
plt.show()
```

<b>Exercise 4

```python
sns.pairplot(df, hue = 'Age', vars = ["Age", "CGPA", "Anxiety (Yes=1, No=0)", "Panic Attack (Yes=1, No=0)"])
#sns.pairplot(df, hue = 'Age')
plt.show()
```

<b>Exercise 5

```python
cgpa_map = {'3.00 - 3.49': 3.25, '3.50 - 4.00': 3.75, '3.50 - 4.00 ': 3.75, '2.50 - 2.99': 2.75, '2.00 - 2.49': 2.25, '0 - 1.99': 1.5}
df['CGPA'] = df['What is your CGPA?'].replace(cgpa_map)

# Convert Yes/No to binary (0/1)
df['Anxiety (Yes=1, No=0)'] = df['Do you have Anxiety?'].map({'Yes': 1, 'No': 0})
df['Panic Attack (Yes=1, No=0)'] = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})

# Select the relevant columns
selected_columns = ['Age', 'CGPA', 'Anxiety (Yes=1, No=0)', 'Panic Attack (Yes=1, No=0)']
df_selected = df[selected_columns]

# Calculate correlation matrix
correlation_matrix = df_selected.corr()

# Create heatmap with annotations
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix - Student Age, CGPA, and Mental Health Status")
plt.show()
```

<b> Exercise 6

```python
# Create a FacetGrid with Depression as the hue variable
g = sns.FacetGrid(df, col="Do you have Depression?", col_wrap=2)

# Map a histogram plot to each facet
g.map(sns.histplot, "CGPA")

# Customize the appearance
g.fig.suptitle("Distribution of CGPA by Depression Status")
g.set_axis_labels("CGPA", "Number of Students")
g.add_legend(title="Depression Status")
plt.show()
```
