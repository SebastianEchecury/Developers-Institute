```python
df2 = titanic_df
df2 = df2.drop(['Cabin'], axis=1)

df2['Age'] = df2['Age'].fillna(df2['Age'].mean())

df2.dropna(subset = ['Embarked'], inplace=True)

print(df2.isnull().sum())
```
