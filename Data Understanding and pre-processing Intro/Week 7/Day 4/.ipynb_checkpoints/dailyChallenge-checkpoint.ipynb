```python
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import os
import random
```

```python
dir = "C:\\Users\\Lenovo\\Desktop\\Developers Institute\\Curso\\Data Understanding and pre-processing Intro\\Week 7\\Day 4\\flowers\\flowers"

files = os.listdir(ruta_directorio)
randomFiles = random.sample(files, 3)
images = list()

for f in randomFiles:
    images.append(dir + '\\' + f)

img1 = Image.open(images[0])
img2 = Image.open(images[1])
img3 = Image.open(images[2])
```

```python
imageRotated = img1.rotate(90) 
imgFliped = img2.rotate(180) 
imgResized = img3.resize((int(img3.width*1.2), int(img3.height*1.2)))
```

```python
def show_images(original, augmented, title):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(augmented)
    plt.title(title)
    plt.axis('off')

    plt.show()
```

```python
show_images(img1, imageRotated, '90 Rotated Image')
```

```python
show_images(img2, imgFliped, 'Fliped Image')
```

```python
show_images(img3, imgResized, 'Resized Image')
```
